```python
import findspark
findspark.init()

import pandas as pd
import pyspark
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .master("local[16]")\
        .appName('PySpark_Df')\
        .getOrCreate()

sc = spark.sparkContext.getOrCreate()
```

```python
## Importa el csv de "data/WorldCupPlayers.csv"
## Visualiza los datos
df = spark.read.csv("./data/WorldCupPlayers.csv",sep=",", header=True, inferSchema=True)
df.show()
```

```python

# Pandas Version to load and make a dataframe and get an rdd
players = pd.read_csv("./data/WorldCupPlayers.csv")
print(players)
# Create a dataframe with the pandas dataframe
df1 = spark.createDataFrame(players)
df1.show()
# the .rdd converts it to a rdd from a dataframe
players_rdd = df.rdd
# Create a SparkContext object
sc = spark.sparkContext.getOrCreate()
# Parallelize the RDD
players_rdd = sc.parallelize(players_rdd.collect())
players_rdd.collect()
```

```python
# Crear una vista temporal
df.createOrReplaceTempView("player")
```

```python
# Replace method in sql could help
spark.sql("select replace(Event, '\\'','') from player").show()
```

```python
# Using like to get all the possible G40 events
spark.sql("select * from player where event like '%G40%' and position = 'C'").show()
spark.sql('''
          select count(1) as total_C_G40 
          from player 
          where event like '%G40%' and position = 'C'
          ''').show()
```

```python
## Utiliza Spark SQL para mostras los registros donde el MatchID sea mayor o igual a 20
spark.sql('''
          SELECT * 
          FROM PLAYER
          WHERE MATCHID >=20
          ''').show()
spark.sql('''
          SELECT COUNT(1) AS AMOUNT_OF_MATCHES_ABOVE20
          FROM PLAYER
          WHERE MATCHID >=20
          ''').show()
```
